## Problem 1-1. First & Second derivative of $f(x)$

**Implement f(x), f_prime(x), f_2prime(x)** functions that return the value of $f(x), f'(x), f''(x)$ for all $x \in \mathbb{R}$.

```python
# Write code here


def f(x):
    
    
    return (x-2)**3*(x-6)+4

def f_prime(x):
    
    
    return 3*(x-2)**2*(x-6)+(x-2)**3

def f_2prime(x):
    
    
    return 6*(x-2)*(x-6)+3*(x-2)**2+3*(x-2)**2
```

```python
# Do Not Modify!!

print(f'f(6) = {f(6)}')
print(f'f\'(4) = {f_prime(4)}')
print(f'f\"(3) = {f_2prime(3)}')
```

## Problem 1-2. Gradient Descent Method

**Implement GD_update** function that updates the parameter estimate using **gradient descent method.**

- Input : 
1. x_old : old parameter estimate  ($x_{t-1}$)
2. learning_rate ($\gamma$)
- Output : 
1. x_new : updated paramter estimate($x_{t}$) 

```python
def GD_update(x_old, learning_rate):
    # Write code here
    x_new = x_old - learning_rate * f_prime(x_old)
    return x_new
```

## Problem 1-3. Newton-Raphson Method.

**Implement NR_update** function that updates the parameter estimate using **Newton-Raphson method.**

* Input:
1. x_old : old parameter estimate($x_{t-1}$)

* Output:
2. x_new : new parameter estimate($x_{t}$)

```python
def NR_update(x_old):
    # Write code here
    return x_old - f_prime(x_old) / f_2prime(x_old)
```

## Problem 1-4. Find the value $x^*$ using gradient descent method.

Implement **GD_optimize** function that satisfies the following conditions.
* Input:<br/>
1) x_zero: initial parameter estimate ($x_0$)<br/>
2) learning rate<br/>
3) threshold
* Output:<br/>
1) params (list that contains parameter estimates $x_0$, $x_1$, $\cdots$, where last element is $x^*$)<br/>
2) values (list that contains values of $f(x_0)$, $f(x_1)$, $\cdots$, where last element is the minimum value, $f(x^*)$)

* How to Implement
1. We want to find the minimizer of $f(x)$ using gradient descent method.
2. Initialize with any $x_0 \in \mathbb{R}.$
3. For each step, <br/>
1) Update $x_{t-1}$ to $x_{t}$ using gradient descent method.<br/>
2) Print the $x_{t}$ and $f(x_{t})$ <br/>
3) Append the $x_{t}$ to "params" list <br/>
4) Append the $f(x_{t})$ to "values" list
4. Repeat step 3 until $|x_{t}-x_{t-1}| <$ threshold. 
5. Print $x^*$ and $f(x^*)$ when $x_t$ has converged.


```python
def GD_optimize(x_zero, learning_rate, threshold):
    # Initialize
    params = [x_zero]
    values = [f(x_zero)]
    # Write code here
    
    x_old = x_zero
    x_new = GD_update(x_old, learning_rate)
    i = 0
    print(f"x_{i} is {x_old}, f(x_{i}) is {f(x_old)}")
    i+=1
    while abs(x_new - x_old) >= threshold:
        
        x_old = x_new
        x_new = GD_update(x_old, learning_rate)
        
        params.append(x_old)
        values.append(f(x_old))
        
        print(f"x_{i} is {x_old}, f(x_{i}) is {f(x_old)}")
        i+=1
    print(f"x_{i} is {x_new}, f(x_{i}) is {f(x_new)}")
    params.append(x_new)
    values.append(f(x_new))
    print(f"x_star is {x_new}, f(x_star) is {f(x_new)}")
    
    return params, values
```

## Problem 1-5. Find $x^*$ using Newton-Raphson method. 

Implement **NR_optimize** function that satisfies the following conditions.
* Input:<br/>
1) x_zero: initial parameter estimate($x_0$)<br/>
2) threshold
* Output:<br/>
1) params (list that contains parameter estimates $x_0$, $x_1$, $\cdots$, where last element is $x^*$)<br/>
2) values (list that contains values of $f(x_0)$, $f(x_1)$, $\cdots$, where last element is minimum value, $f(x^*)$)

* How to Implement
1. We want to find the minimizer of $f(x)$ using Newton-Raphson method.
2. Initialize with any $x_0 \in \mathbb{R}.$
3. For each step, <br/>
1) Update $x_{t-1}$ to $x_{t}$ using Newton-Raphson method.<br/>
2) Print the $x_{t}$ and $f(x_{t})$ <br/>
3) Append the $x_{t}$ to "params" list <br/>
4) Append the $f(x_{t})$ to "values" list
4. Repeat step 3 until $|x_{t}-x_{t-1}| <$ threshold. 
5. Print $x^*$ and $f(x^*)$ when $x_t$ has converged.

```python
def NR_optimize(x_zero, threshold):
    # Initialize
    params = [x_zero]
    values = [f(x_zero)]
    # Write code here:
    x_old = x_zero
    x_new = NR_update(x_old) # Initialize x_new with a value greater than threshold
    
    i = 0
    print(f"x_{i} is {x_old}, f(x_{i}) is {f(x_old)}")
    i+=1
    
    while abs(x_new - x_old) >= threshold:
        x_old = x_new
        x_new = NR_update(x_old)  # Update parameter estimate using Newton-Raphson method
        
        params.append(x_old)
        values.append(f(x_old))
         
        print(f"x_{i} is {x_old}, f(x_{i}) is {f(x_old)}")
        i+=1
    params.append(x_new)
    values.append(f(x_new))
        
    print(f"x_{i} is {x_new}, f(x_{i}) is {f(x_new)}")
    print(f"x_star is {x_new}, f(x_star) is {f(x_new)}")
    
    return params, values
```

## Problem 2-1. Sigmoid
**Implement sigmoid** function.


$\pi(x_i) = \dfrac{exp(\beta_0+\beta_1x_i)}{1+exp(\beta_0+\beta_1x_i)}$

* Input:
1. x_i : $x_i$
2. beta_0 : $\beta_0$
3. beta_1 : $\beta_1$

* Output:
$\pi(x_i)$

```python
def sigmoid(x_i, beta_0, beta_1):
    # Write code here
    exponent = beta_0 + beta_1 * x_i
    return np.exp(exponent) / (1 + np.exp(exponent))
```

## Problem 2-2. log-Likelihood($\beta$)

* Logistic Regression Problem:<br/>
$y_i \sim Ber(\pi(x_i))$, $\pi(x_i) = \dfrac{exp(\beta_0+\beta_1x_i)}{1+exp(\beta_0+\beta_1x_i)}$
<br/><br/>
Hence, $y_i$ is a binary variable taking values 0 or 1 with $P(y_i=1|x_i) = \pi(x_i), P(y_i=0|x_i) = 1-\pi(x_i)$<br/> and probability mass function of $y_i$ is <br/>
$p(y_i|x_i) = \pi(x_i)^{y_i} \times (1-\pi(x_i))^{1-y_i}$
<br/>
Suppose that $y_1, y_2, ... , y_N$ are independent. Then the joint probability mass function of $y_1, y_2, ... , y_N$ is <br/>
$p(y_1, ... , y_N|x_1, ... , x_N) = \prod_{i=1}^{N} \pi(x_i)^{y_i} \times (1-\pi(x_i))^{1-y_i}$.<br/>
==> $Likelihood(\beta)=L(\beta)= \prod_{i=1}^{N} \pi(x_i)^{y_i} \times (1-\pi(x_i))^{1-y_i}$<br/>
==> $Log-likelihood(\beta) = logL(\beta) = l(\beta)$






**Implement log_likelihood** function that calculates the log_likelihood($\beta$) using **sigmoid** function in Problem 2-1.

* Input: 
1. beta_0, beta_1 : $\beta_0, \beta_1$
2. x : $\mathbf{X} \in \mathbb{R}^{N\times1}$
3. y : $\mathbf{y} \in \mathbb{R}^{N\times1}$
* Output:
value of logLikelihood($\beta$), $l(\beta) \in \mathbb{R}$

```python
def log_likelihood(x, y, beta_0, beta_1):
    # Write code here
    l_likelihood = 0
    
    for i in range(len(y)):
        pi = sigmoid(x[i], beta_0, beta_1)
        l_likelihood += y[i] * np.log(pi) + (1 - y[i]) * np.log(1 - pi)
    
    return l_likelihood
```

## Problem 2-3. Score function

Score function is first derivative(gradeint) of log-likelihood function. It defined as below.<br/>$S(\beta)\overset{\underset{\mathrm{def}}{}}{=} \dfrac{\partial l(\beta)}{\partial \beta} = \begin{bmatrix}
\dfrac{\partial l(\beta)}{\partial \beta_0} \\
\dfrac{\partial l(\beta)}{\partial \beta_1} \end{bmatrix} = \bigtriangledown l(\beta) \in \mathbb{R^{2\times1}}$

**Implement score_function** that calculate score function $S(\beta)$ vector.

* Input: 
1. beta_0, beta_1 : $\beta_0, \beta_1$
2. x : $\mathbf{X} \in \mathbb{R}^{N\times1}$
3. y : $\mathbf{y} \in \mathbb{R}^{N\times1}$
* Output:
numpy-array of Score function $S(\beta) \in \mathbb{R^{2\times1}}$

```python
def score_function(x, y, beta_0, beta_1):
    # Write code here
    score = np.zeros((2, 1))
    
    for i in range(len(y)):
        pi = sigmoid(x[i], beta_0, beta_1)
        score[0] += y[i] - pi
        score[1] += (y[i] - pi) * x[i]
    
    return score
```

## Problem 2-4. Hessian Matrix

Hessian Matrix of $l(\beta)$ is second derivative matrix. It can be calculated as below.<br/>

$H(l(\beta)) = l''(\beta) = \bigtriangledown^2 l(\beta) = \dfrac{\partial^2 l(\beta)}{\partial \beta \partial \beta^T} = \begin{bmatrix}
\dfrac{\partial^2 l(\beta)}{\partial \beta_0^2} & \dfrac{\partial^2 l(\beta)}{\partial \beta_0 \partial \beta_1}  \\
\dfrac{\partial^2 l(\beta)}{\partial \beta_1 \partial \beta_0} & \dfrac{\partial^2 l(\beta)}{\partial \beta_1^2}  \\
\end{bmatrix} \in \mathbb{R^{2\times2}}$
<br/>
**Implement hessian_matrix** that calculates hessian matrix of log-likelihood $l(\beta)$.

* Input: 
1. beta_0, beta_1 : $\beta_0, \beta_1$
2. x : $\mathbf{X} \in \mathbb{R}^{N\times1}$
3. y : $\mathbf{y} \in \mathbb{R}^{N\times1}$
* Output:
Hessian matrix of log-likelihood $l(\beta), i.e. H(l(\beta)) \in \mathbb{R^{2\times2}}$ (numpy-array)

```python
def hessian_matrix(x, y, beta_0, beta_1):
    #Write code here
    hessian = np.zeros((2, 2))  # Initialize the Hessian matrix
    
    for i in range(len(y)):
        pi = sigmoid(x[i], beta_0, beta_1)  # Calculate the sigmoid value using the sigmoid function
        hessian[0, 0] -= pi * (1 - pi)
        hessian[0, 1] -= pi * (1 - pi) * x[i]
        hessian[1, 0] -= pi * (1 - pi) * x[i]
        hessian[1, 1] -= pi * (1 - pi) * x[i] * x[i]
    
    return hessian
```

## Problem 2-5. MLE of Logistic Regression parameters $\beta_0, \beta_1$.

**Implement MLE** function that estimates Logistic Regression parameters $\beta_0, \beta_1$ <br/> using **Newton-Raphson methods** and functions in Problem 2-1 ~ 2-4.

* Input:
1. x : $\mathbf{X} \in \mathbb{R}^{N\times1}$
2. y : $\mathbf{y} \in \mathbb{R}^{N\times1}$
3. epochs : number of iterations

* Output:
1. params (list that contains parameters $\beta^{(0)}$, $\beta^{(1)}$, $\cdots$, where last element is true parameter, $\beta^*$)
2. values (list that contains values of log-likelihood $l(\beta^{(0)})$, $l(\beta^{(1)})$, $\cdots$, where last element is  $l(\beta^*)$

* How to Implement
1. We want to find the minimizer of $-l(\beta)$ **using Newton-Raphson method.**
2. Initialize with $\beta^{(0)} = [0,0]^T \in \mathbb{R^{2 \times 1}}.$
3. For each epoch($k$), where $k$ $\in$ (0, 1, ... epochs-1)<br/>
1) Update $\beta^{(k)}$ to $\beta^{(k+1)}$ using Newton-Raphson method.<br/>
2) Print the $\beta^{(k+1)}$ and $l(\beta^{(k+1)})$ <br/>
3) Append the $\beta^{(k+1)}$ to "params" list <br/>
4) Append the $l(\beta^{(k+1)})$ to "values" list
5. Print $\beta^{(epochs)}$ and $l(\beta^{(epochs)})$.  (This should be close to $\beta^*$ and $l(\beta^*)$.)

```python
def MLE(x, y, epochs):
    #Initialize
    params = []
    values = []
    beta_k = np.zeros(2).reshape(-1,1) # Initial beta_0, beta_1 = 0, 0
    # Write code here
    i = 0
    params.append(beta_k) 
    values.append(log_likelihood(x, y, beta_k[0, 0], beta_k[1, 0]))
    print(f"beta_0^({i}) is {beta_k[0, 0]}, beta_1^({i}) is {beta_k[1, 0]}, L(beta) is {values[-1]}")
    i+=1
    for epoch in range(epochs-1):
        score = score_function(x, y, beta_k[0, 0], beta_k[1, 0])
        hessian = hessian_matrix(x, y, beta_k[0, 0], beta_k[1, 0])
        
        beta_k = beta_k - np.dot(np.linalg.inv(hessian), score)
        
        params.append(beta_k) 
        values.append(log_likelihood(x, y, beta_k[0, 0], beta_k[1, 0]))
        
        print(f"beta_0^({i}) is {beta_k[0, 0]}, beta_1^({i}) is {beta_k[1, 0]}, L(beta) is {values[-1]}")
        i+=1
    
    print(f"beta_0_star is {beta_k[0, 0]}, beta_1_star is {beta_k[1, 0]}, L(beta_star) is {values[-1]}")
    return params, values
```
